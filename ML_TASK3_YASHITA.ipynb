```python
#Find the pie chart for color and cut as well so that we are able to analyse which diamond is common and which is rare
```

```python
count_clarity = diamonds_copy['clarity'].value_counts()
```

```python
count=0
for i in count_clarity:
  count+=i
```

```python
percentage_clarity = count_clarity.apply(lambda x: x*100/count)
percentage_clarity
```

```python
colors = sns.color_palette('bright')[0:8]
plt.pie(percentage_clarity.values, labels = percentage_clarity.index, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
count_cut = diamonds_copy['cut'].value_counts()
count=0
for i in count_cut :
  count+=i

percentage_cut = count_cut.apply(lambda x: x*100/count)
percentage_cut
```

```python
colors = sns.color_palette('bright')[0:8]
plt.pie(percentage_cut.values, labels = percentage_cut.index, colors = colors, autopct='%.0f%%')
plt.show()
```
